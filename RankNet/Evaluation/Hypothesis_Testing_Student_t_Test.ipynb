## Hypothesis Testing: Student's T- Test

```python
#STUDENT'S T TEST:
# Hypothesis H0 - two models are identical.
# reject null hypothesis if value of test statistic t > = critical value (cv)
# if Student_t output is:
# a) True - Hypothesis H0 is rejected
# b) False - Hypothesis H0 is correct

def student_t (mod2,mod1):
    difference = mod2 - mod1
    mean = np.mean(difference, axis=0)
    std_dev = np.std(difference,axis = 0)
    N = len(difference)
    sq_root_N = math.sqrt(N)
    if std_dev == 0:
        std_dev = 1
    else:
            std_dev=std_dev       
    t = (mean/std_dev)*sq_root_N
    alpha = 0.05
    """ http://www.sjsu.edu/faculty/gerstman/StatPrimer/t-table.pdf"""
    cv = 1.960
    #reject H0 if t >= cv
    H0_reject = t>=cv
    return H0_reject
```

```python
# Model 1 - Model 13,2,4,8 Comparison
Model1 = test_model_comp['Model 1']
Model13 = test_model_comp['Model 13']
Model2 = test_model_comp['Model 2']
Model4 = test_model_comp['Model 4']
Model8 = test_model_comp['Model 8']

mod13_mod1 = student_t (Model13,Model1)
mod2_mod1 = student_t (Model2,Model1)
mod4_mod1 = student_t (Model4,Model1)
mod8_mod1 = student_t (Model8,Model1)
mod13_mod1, mod2_mod1,mod4_mod1, mod8_mod1
```

```python
# Model 13 - Model 2,4,8 Comparison
mod2_mod13 = student_t (Model2,Model13)
mod4_mod13 = student_t (Model4,Model13)
mod8_mod13 = student_t (Model8,Model13)
mod2_mod13, mod4_mod13, mod8_mod13
```
